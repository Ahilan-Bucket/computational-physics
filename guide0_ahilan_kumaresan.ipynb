## Lennard-Jones Potential
### Task 9 

**Lennard-Jones (LJ) potential**
also known as the Lennard-Jones 12-6 potential is a intermolecular potential Pair (Function) that describes the potential energy of two interacting objects only based on the distance between the two functions. 


$$ V(r) = 4\epsilon [(\frac{\sigma}{r})^{12} - (\frac{\sigma}{r})^6]$$

**Parameters**
- $r$: separation distance between the two particles (length)
- $\epsilon$: depth of the potential well (energy scale of attraction)
- $\sigma$: characteristic length scale at which $V(r)=0$

**Units**
- $r, \sigma$: length (commonly nanometers or Angstrom)
- $V(r), \epsilon$: energy (commonly electronvolt or thermal energy units $k_B T$)

*References*
- https://en.wikipedia.org/wiki/Lennard-Jones_potential  
- https://en.wikipedia.org/wiki/Pair_potential

*"For molecular systems at room temperature, a better
energy scale to use is thermal energy, kBT = 0.025 eV, where kB is Boltzmann’s constant and T=293 K. So
instead of expressing energies in terms of eV, one typically writes energetic parameter in units of
thermal energy for molecular systems at or near room temperature."*

````python
# 4. 
e = 0.12 # ev

# to Convert to KbT. Note that 
KbT = 0.025 #ev

e_to_KbT = e / KbT

print(f"This is now in {e_to_KbT}")
````

````python
# 5. & 6. 
def LJ(rs,epsilon=1,sigma=0.5):
    V = 4*epsilon*((sigma/rs)**12 - (sigma/rs)**6)

    return V
````

Yes, it returns an array of values

````python
# 7. 
r = np.linspace(0.3,0.6,2000)

# 8. 
y = LJ(r,epsilon=1)

# 9.
plt.plot(r,LJ(r,epsilon=1),label=r"$\epsilon$=1")

# 10. & 11. 
plt.plot(r,LJ(r,epsilon=2),label=r"$\epsilon$=2")
plt.plot(r,LJ(r,epsilon=3),label=r"$\epsilon$=3")
plt.plot(r,LJ(r,epsilon=4),label=r"$\epsilon$=4")
plt.plot(r,LJ(r,epsilon=5),label=r"$\epsilon$=5")
plt.plot(r,LJ(r,epsilon=6),label=r"$\epsilon$=6")

# 12. 
plt.legend()
plt.xlabel(r"Distance $r$ ($\sigma$)")
plt.ylabel(r"Potential $V$ ($\epsilon$)")
plt.title("Lennard-Jones Potential")
#plt.loglog()

## 13. Downlode & 14.

plt.savefig("LJ-Plot")
plt.show()

r[np.argmin(y)], np.min(y)

````

## Trajectory Analysis
### Task 10

Notes:
x,y - in Meters of robot

time stamp is 60 Snap shorts per minute (60s)

so 1 snapshot per second.


````python
#? np.genfromtxt

file=r"droid_traj.csv"
data = np.genfromtxt(fname=file,delimiter=",")


x_vals = data[:, 0]
y_vals = data[:, 1]
print(data[:10])


print(data.shape)


#plt.scatter(x=data[:][0],y=data[:][1]), this is incorrect. 
# THis return and entier array and you select 0th or 1st. 

# We need all rows, data[:, __] and then each row, column 0 and 1 data[:, 0/1]


plt.scatter(x=x_vals,y=y_vals)
plt.title("Droid Trajectory Position")
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.grid(color='g')
````

````python
data


# To do this, what I need is, to each state in posisiton, I want to add the 
# cummilative differences. 

# Prof Comments: s(t) to be the cumulative distance traveled, 
# so it should be quite distinct from (the numerical estimate of) the instantaneous velocity.  

# Find the distance between each point
# Then loop through that array and add it to the s_t array one by one cummilatively

# 8.
distance_between_points = np.sqrt(np.diff(x_vals)**2 + np.diff(y_vals)**2)
# Note, an Array of 5 Points has 4 distances

s_t = np.array([0])
running = 0
for i in distance_between_points:
    running += i
    s_t = np.append(s_t,running)

print(s_t[:5],s_t.shape)

````

````python
print(f"Total Distance is: {s_t[-1]} m ")
````

````python
# For Velocity ds/dt, this is velocty travlled for ds (a finite distance) amd dt (finite time) :
dt = 1 # Second
ds = distance_between_points

v_t = ds / dt


# Tangential Acceleration

a_t = np.diff(v_t)/dt
````

### Making Subplots

````python
# For the series of time additions:
t = np.arange(0,len(s_t),1)

plt.subplot(3, 1, 1) # Index starts from 1
plt.plot(t,s_t)
plt.ylabel("s(t) [m]")

plt.subplot(3, 1, 2) # Index starts from 1
plt.plot(t[:-1],v_t)
plt.ylabel("v(t) [m/s]")

plt.subplot(3, 1, 3) # Index starts from 1
plt.plot(t[:-2],a_t)
plt.xlabel("t [s]")
plt.ylabel("at(t) Tangential [m/s^2]")
plt.title("Droid Trajectory Arclength, Velocity and Acceleration")


plt.tight_layout()
plt.show()
````

````python
# TO do colour Mapping
plt.scatter(x=x_vals[:-2],y=y_vals[:-2],c=a_t,cmap="Blues")
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.title("Droid Trajectory Color Mapped by Acceleration")
#plt.grid(color='g')
plt.show()
````
